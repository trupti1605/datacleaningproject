# Importing important libraries

```python
import pandas as pd
import numpy as np
```

# Importing dataset

```python
df = pd.read_csv('https://raw.githubusercontent.com/realpython/python-data-cleaning/master/Datasets/BL-Flickr-Images-Book.csv')
df.head()
```

When we look at the first five entries using the head() method, we can see that a handful of columns provide ancillary
information that would be helpful to the library but isn’t very descriptive of the books themselves: Edition Statement,
Corporate Author, Corporate Contributors, Former owner, Engraver, Issuance type and Shelfmarks.


# Droping columns

```python
to_drop = ['Edition Statement',
           'Corporate Author',
           'Corporate Contributors',
           'Former owner',
           'Engraver',
           'Contributors',
           'Issuance type',
           'Shelfmarks']
df.drop(to_drop, inplace=True, axis=1)
```

```python
df.head()
```

# Changing the Index of a DataFrame

```python
df['Identifier'].is_unique
```

Let’s replace the existing index with this column using set_index:

```python
df = df.set_index('Identifier')
df.head()
```

Tidying up Fields in the Data


```python
df.get_dtype_counts()
```

One field where it makes sense to enforce a numeric value is the date of publication so that we can do calculations down
the road:

```python
df.loc[1905:, 'Date of Publication'].head(10)
```

# Counclusion drawn

1.Convert date ranges to their “start date”, wherever present: 1860-63; 1839, 38-54
2.Completely remove the dates we are not certain about and replace them with NumPy’s NaN: [1897?]
3.Convert the string nan to NumPy’s NaN value

```python
extr = df['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
extr.head()
```

```python
df['Date of Publication'] = pd.to_numeric(extr)
df['Date of Publication'].dtype
```

```python
df['Date of Publication'].isnull().sum() / len(df)
```

```python
df['Place of Publication'].head(10)
```

```python
pub = df['Place of Publication']
london = pub.str.contains('London')
london[:5]
```

```python
 oxford = pub.str.contains('Oxford')
```

```python
df['Place of Publication'] = np.where(london, 'London',
np.where(oxford, 'Oxford',
pub.str.replace('-', ' ')))
```
